### 2. Для каждой книги посчитайте количество обзоров и среднюю оценку. <a id='2'></a>

```python
sql = ''' 

select t.book_id, t.title,   count(distinct t.review_id) as cnt, avg(t.rating) as mean_rating 
from
    (select b.*, r.rating, re.review_id

    from books b
    left join ratings r on r.book_id=b.book_id
    left join reviews re on re.book_id=b.book_id
    ) t
group by 1, 2
order by 1,2

'''
select(sql)
```

### 3. Определите издательство, которое выпустило наибольшее число книг толще 50 страниц — так вы исключите из анализа брошюры. <a id='3'></a>

```python
sql = ''' 
select t.* from
    (select p.publisher_id, p.publisher, count(distinct b.book_id) as cnt --, count(b.title)

    from books b
    left join publishers p on p.publisher_id=b.publisher_id
    where b.num_pages>50
    group by 1,2
    order by count(b.book_id) desc
    ) t
where t.cnt = (select max(t.cnt) from 
    (select p.publisher_id, p.publisher, count(distinct b.book_id) as cnt --, count(b.title)

    from books b
    left join publishers p on p.publisher_id=b.publisher_id
    where b.num_pages>50
    group by 1,2
    order by count(b.book_id) desc
) t)

'''
select(sql)
```

Лидер - издание Penguin Books, выпустил 42 книги за все время наблюдений

### 4. Определите автора с самой высокой средней оценкой книг — учитывайте только книги с 50 и более оценками. <a id='4'></a>

```python
sql = ''' 
select t.*
from
(select a.author, avg(r.rating) as mean_rating, count(*) as cnt_rating
    from books b
    left join authors a on a.author_id=b.author_id
    left join ratings r on r.book_id=b.book_id
    
group by 1
) t
where t.cnt_rating>50 and t.mean_rating = (select max(t.mean_rating) from
(select t.*
from
(select a.author, avg(r.rating) as mean_rating, count(*) as cnt_rating
    from books b
    left join authors a on a.author_id=b.author_id
    left join ratings r on r.book_id=b.book_id
    
group by 1
) t
where t.cnt_rating>50
) t)
'''
select(sql)
```

### 5. Посчитайте среднее количество обзоров от пользователей, которые поставили больше 50 оценок. <a id='5'></a>

```python
sql = '''
select avg(t.count) from (
select t.username, count(distinct t.review_id) from (
select distinct t.username, re.review_id from (
select r.username, count(distinct r.rating_id) as cnt_rating
from ratings r
group by 1
) t
left join reviews re on re.username=t.username
where t.cnt_rating > 50
) t
group by 1
) t
'''
select(sql)
```

В среднем самые активные обзорщики выставляют 24 обзора
